```python
data={'study_hrs':range(1,11),
     'marks obtained':[25,33,46,50,55,67,71,74,80,83]}
df=pd.DataFrame(data)
print(df)
```

```python
# # Y = mx + c
# # M : Slope
# # C : Intercept

# Y = Marks
# x = Study_hrs
```

```python
# x(1), y(1): simple Linear Regression
# x(multiple): y(1): Multiple Linear Regression

df
```

```python
# Import Module
from sklearn.linear_model import LinearRegression

# Object Model Creation
model = LinearRegression()


# Training/learning
x=df[['study_hrs']]
y=df['marks obtained']

model.fit(x,y)
```

```python
x=[[5,5]]
m=model.coef_
c=model.intercept_
print(m,c)
```

```python
std_hrs=[[5.5]]
marks=m*std_hrs+c
print('predicted marks:',marks[0][0])
```

```python
import warnings
warnings.filterwarnings('ignore')
study_hrs=input("enter study hours")
std_hr=[[float(study_hrs)]]
ans=model.predict(std_hr)
marks=round(ans[0],2)
if marks>100:
    marks=100

elif marks<0:
    marks=0

else:
    marks=marks
print(f'your predicted marks is:{marks}')
```
