```python
# Internet Required
df_titanic=sns.load_dataset('titanic')
```

```python
df_titanic.sample()
```

```python
# sns.countplot
# sns.barplot
# sns.boxplot
# sns.scatterplot
# sns.catplot
# sns.pairplot
# sns.heatmap
# sns.regplot
# sns.lmplot
# sns.lineplot
# sns.kdeplot
# sns.violinplot
```

```python
import matplotlib.pyplot as plt
sns.countplot(data = df_titanic, x = 'sex',hue="alive",palette=sns.color_palette("magma",2)) #hue:further division based on categorical data
plt.title('Titanic survival Analysis Bases on gender!')
plt.show()
```

```python
#barlot
sns.barplot(data = df_titanic, x = 'class',y="fare",hue="alive",estimator="mean",palette=sns.color_palette("magma",2))
#confidence interval:ci(95%)
```

```python
sns.barplot(data = df_titanic, x = 'class',y="fare",hue="alive",errorbar=("ci",0))
plt.grid(color = 'r',alpha = 0.1)
plt.show()
```

```python
#boxplot:check distribution of data as well as identifying outliers/fliers:false liers
```

```python
# # sns.boxplot: To detect Outliers and to check data distribution
# # To detect Outliers:
# # Range: Min to max(19 - 22): 35, 28, 45(Outliers: fliers

# # Q2: Quantile : 50%
# # Q1: 25%
# # Q3: 75%

# # IQR = Q3 - Q1
# UW(Upper cap) = Q3 + 1.5*IQR
# LW(Lower cap) = Q1 - 1.5* IQR
```

```python
sns.boxplot(data = df_titanic, x='class', y="fare",hue="sex")
plt.show()
```

```python
sns.boxplot(data = df_titanic, y='class', x="fare",hue="sex",orient='h')
plt.show()
```

```python
# sns.violin plot

plt.figure(figsize = (10,5))
sns.violinplot(data = df_titanic, x='class', y   = 'fare' , hue = 'sex',orient = 'v')
plt.show()
```

```python
df_tips=sns.load_dataset('tips')
df_tips.sample()
```

```python
df_iris=sns.load_dataset('iris')
```

```python
df_iris.iloc[:,:-1]
```

```python
df_iris.iloc[:,:-1].corr()
```

```python
# # >0: +ve : + ve Corr
# =0: No corr
# <0: -ve: - ve Corr
```

```python
sns.pairplot(data=df_iris,hue='species',corner=True)
```

```python
# sns.kdeplot : Kernel density estimation

sns.kdeplot(data = df_iris, x = 'sepal_length', y = 'sepal_width',hue='species')
sns.scatterplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
plt.show()
```

```python
# sns.kdeplot : Kernel density estimation

sns.jointplot(data = df_iris, x = 'sepal_length', y = 'sepal_width',kind = 'hex') # kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
plt.show()
```
